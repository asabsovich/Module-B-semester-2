# 2. Weekly graph question

Below are the graphs from last week's weekly graph question.  Suppose you wanted to cause the viewer to examine the last three months of data first.  How could you use preattentive attributes to do that?  Suggest several possibilities.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size=numdates) / 5 + np.arange(numdates) / 16)
daterange = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')

# Create a DataFrame
df = pd.DataFrame({'date': daterange, 'value': time_series})
df['highlight'] = df['date'] >= '2023-10-01'

# Plot with emphasis on last 3 months
plt.figure(figsize=(10, 5))
plt.plot(df['date'], df['value'], color='grey', linewidth=2)
plt.plot(df[df['highlight']]['date'], df[df['highlight']]['value'], color='red', linewidth=2)
plt.xlabel("Year and month")
plt.ylabel("Nitrate level (% of eutrophic value)")
plt.title("Highlighting Last Three Months")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 99-119 of the Storytelling With Data book as best you can. (The first half of chapter four). You do not have to get the exact data values right, just the overall look and feel.

```python
import matplotlib.pyplot as plt

# Dummy data
categories = ['A', 'B', 'C', 'D', 'E', 'F']
values = [25, 30, 22, 45, 20, 28]

# Emphasis: I want to highlight 'D'
colors = ['gray' if cat != 'D' else 'dodgerblue' for cat in categories]


plt.figure(figsize=(8, 5))
plt.bar(categories, values, color=colors)
plt.title('Category D Stands Out')
plt.xlabel('Category')
plt.ylabel('Value')
plt.tight_layout()
plt.show()
```

This plot reproduces a bar chart from Chapter 4 (p. 100–110) where one bar is highlighted using a bright color (dodgerblue) while the others are muted (gray). This technique draws immediate attention to a specific data point (Category D), leveraging preattentive attributes to guide interpretation and support a focused narrative.
